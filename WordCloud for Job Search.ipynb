```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
file_content=open ("C:/Sumit/scripts/LinkedIn/input/LinkedIn_jobs.txt").read()
```

```python
wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\Verdana.ttf',
                            stopwords=STOPWORDS,
                            background_color = 'white',
                            width=1200,
                            height=1000
                            ).generate(file_content)
```

```python
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
wordcloud.to_file('C:\Sumit\scripts\LinkedIn\output\wordcloud.png')
```
